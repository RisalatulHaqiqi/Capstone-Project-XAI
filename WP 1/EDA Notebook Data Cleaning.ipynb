```python
#Menghapus baris yang mengandung data hilang
df_akhir=df.dropna()
df_akhir
```

```python
#Encode data kategorikal
from sklearn.preprocessing import LabelEncoder

#Menduplikat dataframe untuk encoding
df_encoded = df_akhir.copy()

#Endcode kolom kategorikal
label_encoder = LabelEncoder()
kategori_kolom = ['gender', 'ever_married', 'work_type', 'Residence_type', 'smoking_status']

for col in kategori_kolom:
  df_encoded[col] = label_encoder.fit_transform(df_encoded[col].astype(str))
```

```python
#Korelasi dengan matriks
korelasi_matriks=df_encoded.corr()
print("Korelasi Matriks : ")
print(korelasi_matriks)
```

```python
#Visualisasi Data
import matplotlib.pyplot as plt #Pustaka untuk visualisasi data
import seaborn as sns           #Pustaka untuk visualisasi data

#Output dari visualisasi data akan menjadi notebook
%matplotlib inline
```

```python
#Visualisasi Heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(df_encoded.corr(), annot=True, cmap='RdYlBu', center=0, fmt='.2f')
plt.title('Korelasi Faktor Risiko Stroke\n')
plt.show()
```

```python
#Perbandingan kasus stroke dengan nonstroke
plt.figure(figsize=(6, 4))
df['stroke'].value_counts().plot(kind='bar', color=['skyblue', 'red'])
plt.title('Distribusi Stroke (0=Tidak, 1=Ya)')
plt.xticks(rotation=0)
plt.show()
```

```python
#Visualisasi bar plot untuk hubungan variabel kategorikal dengan stroke. untuk melihat proporsi stroke dalam setiap kategori
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

#Hypertension vs Stroke - apakah hipertensi mempengaruhi risiko stroke?
pd.crosstab(df['hypertension'], df['stroke']).plot(kind='bar', ax=axes[0,0], stacked=True)
axes[0,0].set_title('Hypertension vs Stroke\n(0=Tidak Hipertensi, 1=Hipertensi)')
axes[0,0].legend(['No Stroke', 'Stroke'])

#Heart Disease vs Stroke - apakah penyakit jantung mempengaruhi risiko stroke?
pd.crosstab(df['heart_disease'], df['stroke']).plot(kind='bar', ax=axes[0,1], stacked=True)
axes[0,1].set_title('Heart Disease vs Stroke\n(0=Tidak, 1=Ada)')
axes[0,1].legend(['No Stroke', 'Stroke'])

#Gender vs Stroke - apakah gender mempengaruhi risiko stroke?
pd.crosstab(df['gender'], df['stroke']).plot(kind='bar', ax=axes[1,0], stacked=True)
axes[1,0].set_title('Gender vs Stroke')
axes[1,0].legend(['No Stroke', 'Stroke'])

#Smoking vs Stroke - apakah status merokok mempengaruhi risiko stroke?
pd.crosstab(df['smoking_status'], df['stroke']).plot(kind='bar', ax=axes[1,1], stacked=True)
axes[1,1].set_title('Smoking Status vs Stroke')
axes[1,1].legend(['No Stroke', 'Stroke'])
axes[1,1].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```

```python
#Visualisasi box plot untuk membandingkan distribusi variabel numerik antara stroke dan non-stroke. untuk melihat median, quartile, dan outlier
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

# Age vs Stroke - apakah usia berpengaruh terhadap stroke?
sns.boxplot(x='stroke', y='age', data=df, ax=axes[0], palette=['lightblue', 'salmon'])
axes[0].set_title('Usia vs Stroke\n(0=Tidak Stroke, 1=Stroke)')
axes[0].set_xlabel('Status Stroke')
axes[0].set_ylabel('Usia (tahun)')

# Glucose Level vs Stroke - apakah level glukosa mempengaruhi risiko stroke?
sns.boxplot(x='stroke', y='avg_glucose_level', data=df, ax=axes[1], palette=['lightgreen', 'salmon'])
axes[1].set_title('Glucose Level vs Stroke\n(0=Tidak Stroke, 1=Stroke)')
axes[1].set_xlabel('Status Stroke')
axes[1].set_ylabel('Average Glucose Level')

# BMI vs Stroke - apakah BMI (Indeks Massa Tubuh) berkaitan dengan stroke?
sns.boxplot(x='stroke', y='bmi', data=df, ax=axes[2], palette=['lightyellow', 'salmon'])
axes[2].set_title('BMI vs Stroke\n(0=Tidak Stroke, 1=Stroke)')
axes[2].set_xlabel('Status Stroke')
axes[2].set_ylabel('Body Mass Index')

plt.tight_layout()
plt.show()
```

```python
#Visualisasi Hostogram overlay untuk melihat distribusi variabel numerik dengan overlay stroke dengan non-stroke. Untuk memahami pola distribusi dan perbedaan
fig, axes = plt.subplots(2, 2, figsize=(12, 8))

#Distribusi Usia - distribusi usia antara pasien stroke vs non-stroke
df['age'].hist(bins=30, ax=axes[0,0], alpha=0.6, color='blue', label='Semua pasien')
df[df['stroke']==1]['age'].hist(bins=30, ax=axes[0,0], alpha=0.8, color='red', label='Pasien stroke')
axes[0,0].set_title('Distribusi Usia\n(Merah=Pasien Stroke)')
axes[0,0].set_xlabel('Usia')
axes[0,0].set_ylabel('Frekuensi')
axes[0,0].legend()

#Distribusi Glucose Level - apakah pasien stroke cenderung memiliki glucose lebih tinggi atau tidak
df['avg_glucose_level'].hist(bins=30, ax=axes[0,1], alpha=0.6, color='green', label='Semua Pasien')
df[df['stroke']==1]['avg_glucose_level'].hist(bins=30, ax=axes[0,1], alpha=0.8, color='red', label='Pasien Stroke')
axes[0,1].set_title('Distribusi Glucose Level\n(Merah=Pasien Stroke)')
axes[0,1].set_xlabel('Average Glucose Level')
axes[0,1].set_ylabel('Frekuensi')
axes[0,1].legend()

#Distribusi BMI - apakah ada perbedaan pola BMI antara stroke vs non-stroke?
df['bmi'].hist(bins=30, ax=axes[1,0], alpha=0.6, color='orange', label='Semua Pasien')
df[df['stroke']==1]['bmi'].hist(bins=30, ax=axes[1,0], alpha=0.8, color='red', label='Pasien Stroke')
axes[1,0].set_title('Distribusi BMI\n(Merah=Pasien Stroke)')
axes[1,0].set_xlabel('Body Mass Index')
axes[1,0].set_ylabel('Frekuensi')
axes[1,0].legend()

# Menyembunyikan subplot yang tidak terpakai
axes[1,1].set_visible(False)

plt.tight_layout()
plt.show()
```
